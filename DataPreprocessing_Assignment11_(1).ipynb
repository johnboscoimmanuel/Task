**Step 1: Importing the libraries**

```python

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import warnings
warnings.filterwarnings('ignore')
```

**Step 2: Importing dataset**

```python
df=pd.read_csv('/content/Data.csv')
df.head()
```

**Step 3: Handling the missing data**

```python
df.isnull().sum()
```

```python
df.fillna(df['Age'].mean(),inplace=True)

df.fillna(df['Salary'].mean(),inplace=True)
```

```python
df.isnull().sum()
```

**Step 4: Encoding categorical data**

```python
df['Purchased']=df['Purchased'].map({'Yes':1,'No':0})
```

```python
df
```

**Step 5: Creating a dummy variable**

```python
df_country=pd.get_dummies(df['Country'])
df=pd.concat([df,df_country],axis=1)
df
# x=df.drop(columns=['Age','Salary'],axis=1)
# y=df['Purchased']
```

```python
df.drop('Country',axis=1,inplace=True)
```

```python
df
```

**Step 6: Splitting the datasets into training sets and Test sets**

```python
x_train,x_test=train_test_split(df,test_size=0.2,random_state=2)
```

```python
x_train
```

```python
x_test
```

**Step 7: Feature Scaling**

```python
scaler=StandardScaler()

scaler.fit(x_train,x_test)

stand_scaler=scaler.fit_transform(x_train,x_test)

stand_scaler
```
